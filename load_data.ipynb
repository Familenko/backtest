```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
start = "2023-01-01"
end = "2026-02-01"

symbols = {
    "BTC": "BTC-USD",
    "ETH": "ETH-USD",
    "BNB": "BNB-USD",
    "XRP": "XRP-USD",
    "SOL": "SOL-USD",
    "TRON": "TRX-USD",
    "ADA": "ADA-USD",
    "XMR": "XMR-USD",

    "LINK": "LINK-USD",
    "AVAX": "AVAX-USD",
    "DOT": "DOT-USD",
    "LTC": "LTC-USD"
}
```

```python
df = pd.DataFrame()
for k, v in symbols.items():
    df[k] = yf.download(v, start=start, end=end)["Close"]

df.to_csv("prices.csv")
```
