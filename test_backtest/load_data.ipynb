```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
start = "2023-01-01"
end = "2026-02-01"

symbols = {
    "BTC": "BTC-USD",
    "ETH": "ETH-USD",
    "BNB": "BNB-USD",
    "XRP": "XRP-USD",
    "SOL": "SOL-USD",
    "TRON": "TRX-USD",
    "ADA": "ADA-USD",
    "XMR": "XMR-USD",

    "LINK": "LINK-USD",
    "AVAX": "AVAX-USD",
    "DOT": "DOT-USD",
    "LTC": "LTC-USD"
}

df = pd.DataFrame()
for k, v in symbols.items():
    df[k] = yf.download(v, start=start, end=end, auto_adjust=False)["Close"]

df.to_csv("data/crypto.csv")
```

```python
import yfinance as yf
import pandas as pd

start = "2019-01-01"
end = "2026-01-01"

symbols = [
    "HSBA.L",   # Великобританія — HSBC: фінанси, банківські послуги
    "BAB.L",    # Великобританія — Babcock International Group: флот, техобслуговування, військова інфраструктура
    'BP.L',     # Великобританія — BP: енергетика, нафта, газ

    "TSM",       # Тайвань — TSMC: виробництво чипів
    "ASML.AS",   # Європа — ASML: виробництво обладнання для чипів

    "JNJ",       # США — Johnson & Johnson: медицина
    "DIS",       # США — Disney: розваги, медіа
    "GOOGL",     # США — Alphabet: наука, AI
    "SPY"     # США — SPY: індекс S&P 500
]

df = pd.DataFrame()
for ticker in symbols:
    df[ticker] = yf.download(ticker, start=start, end=end, auto_adjust=False)["Close"]

df.to_csv("data/stock.csv")
```

```python
import yfinance as yf
import pandas as pd

start = "2023-01-01"
end = "2026-01-01"

# UCITS ETF з біржовими суфіксами Yahoo Finance
symbols = [
    # ===== США (S&P 500) =====
    "CSPX.L",   # iShares Core S&P 500 UCITS (London Stock Exchange)
    "VUSA.L",   # Vanguard S&P 500 UCITS (LSE, distributing)

    # ===== Весь світ =====
    "IWDA.AS",  # iShares Core MSCI World UCITS (Euronext Amsterdam)
    "VWCE.DE",  # Vanguard FTSE All-World UCITS (Xetra)

    # ===== Європа =====
    "IMEU.L",   # iShares Core MSCI Europe UCITS (LSE)
    "EXSA.DE",  # iShares STOXX Europe 600 UCITS (Xetra)

    # ===== Єврозона =====
    "EUN2.DE",  # iShares EURO STOXX 50 UCITS (Xetra)

    # ===== Emerging Markets =====
    "EIMI.L",   # iShares Core MSCI EM IMI UCITS (LSE)
]
df = pd.DataFrame()

for ticker in symbols:
    data = yf.download(ticker, start=start, end=end, auto_adjust=False)
    df[ticker] = data["Close"]

df.to_csv("data/ucits_etf_market.csv")
```
